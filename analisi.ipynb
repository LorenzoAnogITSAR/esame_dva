```python
# Colonne da controllare
cols = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating']

# Filtra le righe con almeno un valore mancante nelle colonne indicate
righe_vuote = df_pulito[df_pulito[cols].isna().any(axis=1)]

# Raggruppa per anno e conta quante righe mancano
mancanti_per_anno = (
    righe_vuote
    .groupby('Year_of_Release')
    .size()
    .reset_index(name='missing_rows')
)

mancanti_per_anno
```

```python
# Colonne da controllare
cols = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating']

# Filtra le righe con almeno un valore mancante nelle colonne indicate
righe_vuote = df_pulito[df_pulito[cols].isna().any(axis=1)]

# Definisci i bin (fasce di anni)
bins = [-float('inf'), 1999, 2006, 2012, 2015]
labels = ['<2000', '2000-2006', '2007-2012', '2013-2015']

# Crea una nuova colonna con il range di appartenenza
righe_vuote['Year_Range'] = pd.cut(righe_vuote['Year_of_Release'], bins=bins, labels=labels)

# Raggruppa per range e conta le righe
mancanti_per_range = (
    righe_vuote
    .groupby('Year_Range')
    .size()
    .reset_index(name='missing_rows')
)

mancanti_per_range
```
